```
URL = 'https://www.gumtree.com.au/s-garage-sale/garage+sale/k0c18486'
  
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')


mainPage=soup.find("div",{"class":"search-results-page"})
#find the total number of pages

data=mainPage.find('a',class_='page-number-navigation__link page-number-navigation__link-last link link--base-color-primary link--hover-color-none link--no-underline')
lastPageString=data.get('href')
lastPageNumber=int((lastPageString[lastPageString.find('page-')+5:lastPageString.find('page-')+7]))

#show first 3 pages only
lastPageNumber=3

#iterate through Pages
for page_num in range(lastPageNumber):

  URL = 'https://www.gumtree.com.au/s-garage-sale/garage+sale/page-'+str(page_num+1)+ '/k0c18486'
  
  page = requests.get(URL)

  soup = BeautifulSoup(page.content, 'html.parser')


  mainPage=soup.find("div",{"class":"search-results-page"})

  mainPage.find('div',{"class":"search-results-page__result-nav"}).decompose()

  data=json.loads(mainPage.find('script',type='application/ld+json').text)
  #print(data['itemListElement'])

  print("-----------PAGE NUMBER",page_num+1,"----------------")
  print("\n")
#iterate through each ad in page
  for item in data['itemListElement']:
    #print(item['url'])
    URL = item['url']
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    #get location
    results=soup.find('span',class_='vip-ad-title__location-address')
    #get seller name
    sellerName=soup.find('div',class_='seller-profile__detail')
    print(results.text.strip()+"\t"+sellerName.span.span.text)
  print("\n")
  print("-------------END OF PAGE----------------")
  print("\n")
```
